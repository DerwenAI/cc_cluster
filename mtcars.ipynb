Load the library dependencies

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

Load the dataset

```python
url: str = "https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv"
df: pd.DataFrame = pd.read_csv(url)
df = df.set_index("model")
```

```python
df.head()
```

```python
# standardize
sns.clustermap(df, standard_scale = 1)
plt.show()
```

```python
# normalize
sns.clustermap(df, z_score = 1)
plt.show()
```
